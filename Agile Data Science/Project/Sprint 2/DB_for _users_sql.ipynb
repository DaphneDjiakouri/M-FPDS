```python
import pandas as pd
from sqlalchemy import create_engine, Column, Integer, String, MetaData, Table
from sqlalchemy.ext.declarative import declarative_base

# Load data from CSV
csv_file_path = 'names.csv'
df = pd.read_csv(csv_file_path)

# Define SQLAlchemy base
Base = declarative_base()

# Define Users table model
class Users(Base):
    __tablename__ = 'users'
    id = Column(Integer, primary_key=True)
    name = Column(String(255))

# Create SQLite database in memory (you can change this to a different database)
engine = create_engine('sqlite:///:memory:', echo=True)

# Create the table
Base.metadata.create_all(engine)

# Insert data into the table
df.to_sql('users', engine, index_label='id', if_exists='replace')

# Query the database to verify
result = engine.execute("SELECT * FROM users").fetchall()
print(result)

```
